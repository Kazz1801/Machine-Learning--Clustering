```python
import numpy as np, random
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import Birch
from sklearn.datasets import make_blobs
from sklearn import metrics 
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('Mall_Customers.csv').dropna().drop('CustomerID', axis=1)
features = ['Annual Income (k$)', 'Spending Score (1-100)'] 
X = data[features]
```

```python
random.seed(1)
X, _ = make_blobs(n_samples=200, centers=5, cluster_std=1.2)
```

```python
Brc = Birch(branching_factor = 200, n_clusters =3, threshold = 1.5)
Brc.fit(X)
labels = Brc.predict(X)
print("BIRCH Parameters:", Brc.get_params())
plt.scatter(X[:,0], X[:, 1], c = labels, cmap = 'rainbow', alpha = 0.7, edgecolors = 'b')
plt.title("BIRCH Clustering")
plt.show()
```
